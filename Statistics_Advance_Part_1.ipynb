## Q1. What is a random variable in probability theory?

## ANS 1.

- Random variable is a function that assigns values to each of the experimental outcomes.

- It can take any value from the set of values.

- They can be discrete or continuous.

- **Example:**  
Rolling a die: The number of dots on the face of the die is a random variable, which can take on values 1, 2, 3, 4, 5, or 6.

## Q2. What are the types of random variables?

## ANS 2.

Random Variables are of two types:
1. **DISCRETE RANDOM VARIABLE:**  A discrete random variable can take on a countable number of distinct values.   
Examples:
    - Number of heads in a coin flip
    - Number of students in a class

2. **CONTINUOUS RANDOM VARIABLE:**  A continuous random variable can take on any value within a given interval or range.  
Examples:
    - Height of a person
    - Weight of an object

## Q3. What is the difference between discrete and continuous distributions?

## ANS 3.

Here are the differences between discrete and continuous random variables:

1. **Definition:** A discrete random variable can take on a countable number of distinct values whereas a continuous random variable can take on any value within a given interval or range.  

2. **Nature of values:** Discrete random variables have countable values, while continuous random variables have uncountable values.

3. **Number of outcomes**: Discrete random variables have finite or countable outcomes, while continuous random variables have infinite outcomes.

4. **Descriptive function**: Discrete random variables are described by a Probability Mass Function (PMF), whereas continuous random variables are described by a Probability Density Function (PDF).

5. **Examples**: Discrete random variables include the number of heads in coin flips, while continuous random variables include the height of a person.

## Q4. What are probability distribution functions (PDF)?

## ANS 4.

- A function that gives the probability of each possible value of a random variable (discrete or continuous) is called **probability distribution function**.

- There are two types of probability distribution functions:

    1. **Probability Mass Function (PMF):** For discrete random variables, the PMF describes the probability of each distinct value.

    2. **Probability Density Function (PDF):** For continuous random variables, the PDF describes the probability of each value within a given range.

## Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

## ANS 5.

Here are the differences between Cumulative Distribution Functions (CDF) and Probability Distribution Functions (PDF):

1. **Purpose:** While PDF describes the probability of individual values, CDF describes the probability of values up to a point.

2. **Nature of values:** While PDF values represent probabilities of individual values, CDF values represent cumulative probabilities up to a given value.

3. **Continuity:** While PDF can be discrete or continuous, CDF is always continuous.

4. **Interpretation:** While PDF is used to calculate probabilities of specific values, CDF is used to calculate probabilities of values within a range.

5. **Graphical representation:** While PDF graphs show individual probabilities as spikes or curves, CDF graphs show cumulative probabilities as a continuous curve.

## Q7. What are the key properties of a Bernoulli distribution?

## ANS 7.

The key properties of a **Bernoulli distribution** are:

1. **Binary outcome:** Only two possible outcomes (success or failure, 0 or 1)

2. **Fixed probability:** Probability of success ($p$) is fixed and remains the same for each trial

3. **Independent trials:** Each trial is independent of the other

4. **Finite trials:** Number of trials should be finite

4. **Probability mass function:** P(X=1) = $p$,  P(X=0) = $1-p$  

5. **Mean:** $\mu$ = $p$

6. **Variance:** $\sigma^2$ = $p(1-p)$

## Q8. What is the binomial distribution, and how is it used in probability?

## ANS 8.

- **Binomial distribution** is a way to calculate the probability of getting a certain number of successes (or failures) in a fixed number of tries.

- It is the $n$ Bernoulli trial.

- **Characteristics :**  
1. The number of trials is fixed and known.
2. The probability of success $p$ remains the same for each trial.
3. Each trial is independent of the other.
4. Each trial has **only two** possible outcomes: success or failure.
5. Probability mass function:
P(X = $k$) = $^nC_k$   $p^k$ $(1-p)^{n-k}$  
where:
  - $n$ = number of trials
  - $k$ = number of successes
  - $^nC_k$ = number of combinations of $n$ items taken $k$ at a time
  - $p$ = probability of success
  - $1-p$ = probability of failure


- **The Binomial Distribution is used in probability to:**  
  1. Find the probability of getting exactly $k$ successes in $n$ trials.

  2. Analyze events with two outcomes (success/failure, heads/tails, etc.).

  3. Estimate the number of successes in a fixed number of trials.

## Q9. What is the Poisson distribution and where is it applied?

## ANS 9.

The Poisson Distribution is a discrete probability distribution that describes the number of events occurring within a fixed interval of time or space, where these events occur independently and at a constant average rate.

- PMF : P(X=$x$) = $\frac{e^{-\lambda}\lambda ^x}{x!}$  
where:
    - $e$ = 2.71828
    - $\lambda$ = average rate of events every interval
    - $x$ = outcome

- **Applications:**

    1. **Queueing:** Modeling the number of arrivals in a queue (e.g., customers, calls, or packets).

    2. **Insurance:** Modeling the number of claims or accidents.

    3. **Biology:** Modeling the number of cells, organisms, or mutations.

    4. **Telecommunications:** Modeling the number of calls or messages.

## Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

## ANS 13.

**Central Limit Theorem**  states that if we have a population with mean $\mu$ and standard deviation $\sigma$ and take sufficiently large number of samples from the population with replacement, then the distribution of sample means will be approximately normally distributed.  
In other words, sampling mean of population($\mu, \sigma$) will approximately be a normal distribution($\mu, \frac{\sigma}{\sqrt n}$).

- Why is the CLT critical in statistics?

  1. The CLT enables us to make inferences about a population based on a sample.

  2. Many statistical tests, such as t-tests and ANOVA, rely on the CLT.

  3. The CLT allows us to predict the probability of observing a certain sample mean.

  4. The CLT makes it possible to analyze data from a variety of distributions.

## Q14. How does the Central Limit Theorem relate to the normal distribution?

## ANS 14.

The Central Limit Theorem (CLT) relates to the normal distribution in the following ways:

1. Convergence to normality: The CLT states that the distribution of the sample mean will converge to a normal distribution as the sample size increases.

2. Approximation: For large sample sizes, the CLT allows us to approximate the distribution of the sample mean as a normal distribution, even if the underlying population distribution is not normal.


3. Standardization: The CLT enables us to standardize the sample mean by subtracting the population mean and dividing by the standard deviation ($\frac{x - \mu}{\sigma}$), resulting in a standard normal distribution(z-distribution).

## Q16. How do you calculate a Z-score, and what does it represent?

## ANS 16.

- Z-score is calculated by the formula:  
$Z_{score}$ = $\frac{x - \mu}{\sigma}$  
where:
    - $x$ = raw score (data point)
    - $\mu$ = population mean
    - $\sigma$ = population standard deviation  


- What does a Z-score represent?  
A Z-score represents the number of standard deviations away from the population mean a data point is.

- Interpretation:
  - A Z-score of 0 means the data point is equal to the population mean.

  - A positive Z-score means the data point is above the population mean.

  - A negative Z-score means the data point is below the population mean.

  - The magnitude of the Z-score indicates how many standard deviations away from the mean the data point is.

## Q17. What are point estimates and interval estimates in statistics?

## ANS 17.

### **Point Estimate:**
A point estimate is a single value that is used to estimate a population parameter.

- **Example:** Using the sample mean ($\overline x$) to estimate the population mean ($\mu$).
- **Advantages:** Easy to calculate and understand.
- **Disadvantages:** Does not provide a range of values or a measure of uncertainty.

### **Interval Estimate:**
An interval estimate, also known as a confidence interval, is a range of values that is likely to contain the population parameter.

- **Example:** Creating a 95% confidence interval for the population mean ($\mu$) based on the sample mean ($\overline x$) and standard error ($\frac{\sigma}{\sqrt n}$).
- **Advantages:** Provides a range of values and a measure of uncertainty.
- **Disadvantages:** Can be wider or narrower depending on the sample size and confidence level.

## Q18. What is the significance of confidence intervals in statistical analysis?

## ANS 18.

Confidence intervals (CIs) play a crucial role in statistical analysis:

1. Estimation : CIs provide a range of values within which a population parameter is likely to lie.

2. CIs correct the uncertainty associated with a point estimate.

3. CIs indicate the precision of an estimate, with narrower intervals indicating greater precision.

4. CIs enable comparison of estimates across different groups or studies.

5. CIs can be used to test hypotheses by checking if a parameter falls within or outside the interval.

## Q20. How are Z-scores used to compare different distributions?

## ANS 20.

## Q22. What is the concept of expected value in a probability distribution?

## ANS 22.

The Expected Value (EV) of a probability distribution is the long-run average value of a random variable.

- It represents the center or "balance point" of the distribution.

- EV = $\sum x  P(x)$  
where:
     - $x$ = possible values of the random variable
     - $P(x)$ = probability of each random value
     
- **Long-run average:** The EV is the average value one would expect to observe over many trials or observations.

**Example:**

A fair six-sided die:  

EV = (1 * 1/6) + (2 * 1/6) + (3 * 1/6)+ (4 * 1/6)+ (5 * 1/6) + (6 * 1/6) = 3.5

The expected value of rolling a fair die is 3.5, indicating that, on average, one would expect to roll a 3.5 over many trials.

## Q23. How does a probability distribution relate to the expected outcome of a random variable?

## ANS 23.

- A probability distribution defines the likelihood of different outcomes of a random variable. The expected value is the long-term average or mean outcome of the random variable, derived from the probability distribution.

- The expected value depends on the form and parameters of the probability distribution and provides insights into the "center" or average behavior of the random variable.

- **Example with a Discrete Random Variable:**  
Consider a discrete uniform distribution where a six-sided die is rolled. The outcomes are the integers 1 through 6, each with probability $P(x)= \frac{1}{6}$.  
The probability distribution is:  
$P(1)=P(2)=P(3)=P(4)=P(5)=P(6)= \frac{1}{6}$  
The expected value is calculated as:     
$E[X] = 1 . \frac{1}{6} + 2. \frac{1}{6} + 3. \frac{1}{6} + 4. \frac{1}{6} + 5. \frac{1}{6} +6.\frac{1}{6}$ $= \frac{1+2+3+4+5+6}{6} = \frac{21}{6} = 3.5$  
Thus, the expected value of the roll of a fair die is 3.5, which is the long-term average if we repeatedly roll the die many times.

## Q9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
def calculate_and_plot_zscores(data):
    import numpy as np
    import matplotlib.pyplot as plt

    data_mean = np.mean(data)
    data_std = np.std(data)

    z_scores = (data - data_mean) / data_std

    plt.figure(figsize=(8, 6))
    plt.plot(z_scores, marker='o', linestyle='-', markersize=4)
    plt.xlabel("Data Point Index")
    plt.ylabel("Z-score")
    plt.title("Z-scores of the Dataset")
    plt.grid(True)
    plt.show()

    return z_scores
```

## Q10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

size_of_population = 10000
n = 50
num_samples = 1000

population_data = np.random.exponential(scale=2, size=size_of_population)

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population_data, size=n)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

plt.hist(sample_means, bins=30, density=True, edgecolor='black', alpha=0.7)

mean_sample_means = np.mean(sample_means)
std_sample_means = np.std(sample_means)

x = np.linspace(min(sample_means), max(sample_means), 1000)
pdf = (1 / (std_sample_means * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_sample_means) / std_sample_means) ** 2)

plt.plot(x, pdf, color='red', label='Normal Approximation')
plt.xlabel('Sample Means')
plt.ylabel('Density')
plt.title('Central Limit Theorem: Sampling from Exponential Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

## Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

def calculate_zscore(x, mu, sigma):
    return (x - mu) / sigma

def compare_to_standard_normal(z_score):
  p_value = norm.cdf(z_score)

  print(f"Z-score: {z_score}\n")
  print(f"Probability (p-value) of observing a value less than or equal to {z_score}: {p_value:.4f}\n")

  x = np.linspace(-4, 4, 1000)
  y = norm.pdf(x, 0, 1)

  plt.plot(x, y)
  plt.fill_between(x[x<=z_score], 0, norm.pdf(x[x<=z_score], 0, 1), alpha=0.5)
  plt.title('Standard Normal Distribution with Z-score')
  plt.xlabel('Z-score')
  plt.ylabel('Probability Density')
  plt.grid(True)
  plt.show()

data_point = 75
population_mean = 70
population_std_dev = 5

z = calculate_zscore(data_point, population_mean, population_std_dev)
compare_to_standard_normal(z)
```

## Q15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    se = population_std / (sample_size**0.5)

    z_statistic = (sample_mean - population_mean) / se
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    if p_value < alpha:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"

    return z_statistic, p_value, decision

sample_mean = 72
population_mean = 70
population_std = 5
sample_size = 30

z_statistic, p_value, decision = z_test(sample_mean, population_mean, population_std, sample_size)

print(f"Z-statistic: {z_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
decision
```

## Q16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

standard_error = sample_std / np.sqrt(n)
confidence_level = 0.95
alpha = 1 - confidence_level
df = n - 1
critical_value = stats.t.ppf(1 - alpha / 2, df)

margin_of_error = critical_value * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
##Interpretation >>> We are 95% confident that the true population mean lies within the calculated confidence interval (47.16, 50.76).
```

## Q18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 0
sigma = 1
x = np.linspace(-5, 5, 1000)

pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='Normal Distribution PDF (calculated)', color='magenta', lw=3)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

## Q19.  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambda_ = 5
k_values = np.arange(0, 15)
cdf_values = poisson.cdf(k_values, lambda_)

plt.figure(figsize=(8, 6))
plt.step(k_values, cdf_values, where='post', label='Poisson CDF', color='purple', lw=2)
plt.xlabel('Number of Events (k)')
plt.ylabel('Cumulative Probability')
plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (λ={lambda_})')
plt.grid(True)
plt.legend()
plt.show()

for k, cdf_value in zip(k_values, cdf_values):
    print(f"\n P(X <= {k}) = {cdf_value:.4f}")
```

## Q23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
original_data = np.random.normal(loc=50, scale=10, size=1000)

mean = np.mean(original_data)
std_dev = np.std(original_data)

normalized_data = (original_data - mean) / std_dev

plt.figure(figsize=(12, 6))

#Plot original data
plt.subplot(1, 2, 1)
plt.hist(original_data, bins=30, alpha=0.7, color='blue', label='Original Data')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean:.2f}')
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=2, label=f'Standard Deviation = {std_dev:.2f}')
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=2)
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

#Plot normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, alpha=0.7, color='orange', label='Normalized Data')
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean = 0')
plt.axvline(0+1, color='green', linestyle='dashed', linewidth=2, label='Standard Deviation = 1')
plt.axvline(0-1, color='green', linestyle='dashed', linewidth=2)
plt.title('Normalized Data Distribution (Z-score)')
plt.xlabel('Z-score Value')
plt.ylabel('Frequency')

plt.legend()
plt.tight_layout()
plt.show()
```

## Q24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)
    return data_skewness, data_kurtosis

np.random.seed(42)
dataset = np.random.normal(loc=50, scale=10, size=75)

skewness_value, kurtosis_value = calculate_skewness_kurtosis(dataset)

print(dataset)
print("\n")
print(f"Skewness of the dataset: {skewness_value:.4f}\n")
print(f"Kurtosis of the dataset: {kurtosis_value:.4f}\n")
```
